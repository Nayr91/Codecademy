# U.S. Medical Insurance Costs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
insurance = pd.read_csv(r'C:\Users\ryan-\insurance.csv')
```

```python
insurance.head()
```

```python
insurance.describe()
```

```python
insurance.region.mode()
```

```python
smoker_avg = insurance[insurance.smoker == 'yes'].charges.mean()
print(smoker_avg)
```

```python
non_smoker_avg = insurance[insurance.smoker == 'no'].charges.mean()
print(non_smoker_avg)
```

```python
smoker_avg - non_smoker_avg
```

```python
insurance[insurance.children >= 1].age.mean()
```

```python
sns.histplot(x="age", y='charges', data=insurance[insurance.smoker == 'yes'])
plt.show()
plt.clf()
```

```python
sns.histplot(x="age", y='charges', data=insurance[insurance.smoker == 'no'])
plt.show()
plt.clf()
```

```python
children_df = insurance[insurance.children > 1]
sns.histplot(x="age", y='charges', data=children_df[children_df.smoker == 'no'])
plt.show()
plt.clf()
```

```python
no_children_df = insurance[insurance.children <= 1]
sns.histplot(x="age", y='charges', data=no_children_df[no_children_df.smoker == 'no'])
plt.show()
plt.clf()
```

```python
low_bmi = insurance[insurance.bmi < 20]
sns.histplot(x="age", y='charges', data=low_bmi[low_bmi.smoker == 'no'])
plt.show()
plt.clf()
```

```python
high_bmi = insurance[insurance.bmi > 20]
sns.histplot(x="age", y='charges', data=high_bmi[high_bmi.smoker == 'no'])
plt.show()
plt.clf()
```

```python
#conclusion - The heaviest effects on the charges are if you have a high BMI, and/or are a smoker. 
# The number of Children does not seem to have much effect
```
